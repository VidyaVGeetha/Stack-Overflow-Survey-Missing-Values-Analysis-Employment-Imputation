## ___________Missing Values Heatmap — Selected Columns_________________

```python

# Step 6: Heatmap of Missing Values in Key Columns
import matplotlib.pyplot as plt

cols = ["Employment", "RemoteWork", "WorkExp", "ConvertedCompYearly", "JobSat"]

plt.figure(figsize=(12,6))

sns.heatmap(
    df[cols].isnull(),
    cbar=False
)

plt.title("Missing Values Heatmap (Selected Columns)",fontsize=18, fontweight="bold" )
plt.show()
```

### Key Insihgts:
* Employment → almost no missing values.
* RemoteWork → many missing values in different rows.
* WorkExp → several missing values spread across the data.
* ConvertedCompYearly (salary) → a large block of values is missing.
* JobSat → some missing values, but fewer than salary.

# Task 4: Imputing Missing Values

```python
# Step 1: Find the most frequent value (mode) in Employment
majority_value = df["Employment"].mode()[0]
print("Most frequent value in Employment:", majority_value)
```

Most people in the dataset are "full-time employees".

```python
# Step 2: Check how many times it appears

count = df["Employment"].value_counts()[majority_value]
print("Number of times it appears:", count)
```

There are 39,041 people in the dataset who are employed full-time.

# Task 5: Visualizing Imputed Data

```python
# Step 1: Filling Missing Values in the Employment Column Using the Mode

majority_value = df["Employment"].mode()[0]
df["Employment"].fillna(majority_value, inplace=True)
```

```python
# Step 2: Top 5 Employment Categories After Imputation

top5 = df["Employment"].value_counts().head(5)

plt.figure(figsize=(9,5))    # Creates a new figure for the plot

ax = sns.barplot(           # creates a horizontal bar chart
    y=top5.index,
    x=top5.values,
    hue=top5.index,
    palette="rocket",
    legend=False
)

# --- COLORS YOU CAN CHANGE ---
category_color = "#0F4C81"   # y-axis category labels
value_color = "#0B5345"      # numbers on bars

# Color the category names (y-axis tick labels)
plt.yticks(color="#006D5B", fontsize=12)

# Add colored value labels on bars
for bar in ax.patches:
    ax.text(
        bar.get_width() + 500,
        bar.get_y() + bar.get_height()/2,
        f"{int(bar.get_width())}",
        va="center",
        fontsize=11,
        color=value_color
    )

plt.title(
    "Top 5 Employment Categories After Imputation",
    fontsize=20, fontweight="bold", color="#0F4C81"
)

plt.xlabel("Count", fontsize=18, color="#0F4C81")
plt.ylabel("Employment Category", fontsize=18, color="#0F4C81")

plt.grid(axis="x", linestyle="--", alpha=0.3)
sns.despine(left=True, bottom=True)

plt.tight_layout()
plt.show()
```

### Key Insihgts:
* Missing Employment values were filled using the most common value (mode).
* The chart shows that full-time employment is the largest group.
* Other categories appear much less frequently.
* This confirms that using the mode for imputation was appropriate.
